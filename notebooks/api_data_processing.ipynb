Пишем функцию, которая приведет з/п в нормальный вид

```python
def salary_normalizing(matches):
  salaries = []
  for m in matches:
    m = m.strip()
    if any(k in m.lower() for k in ['к', 'k', 'тыс']):
      num = ''.join(c for c in m if c.isdigit())
      if num:
        salaries.append(int(num) * 1000)
    else:
      num = ''.join(c for c in m if c.isdigit())
      if num:
        salaries.append(int(num))
  salaries = [s for s in salaries if 50000 <= s <= 5000000]
  if not salaries:
    return None, None, None
  salary_from = min(salaries)
  salary_to = max(salaries) if len(salaries) > 1 else None
  salary_mid = (salary_from + salary_to) / 2 if salary_to else salary_from
  return salary_from, salary_to, salary_mid
```

```python
matches = df_tg['text'].str.findall(r'\d[\d\s]*[\.\s]\d{3}|\d{4,}|[1-9]\d{1,2}\s?[кkКK]|\d+\s?тыс')
matches
```

```python
results = matches.apply(salary_normalizing)
df_tg['salary_from'] = results.apply(lambda x: x[0])
df_tg['salary_to'] = results.apply(lambda x: x[1])
df_tg['salary_mid'] = results.apply(lambda x: x[2])
```

```python
df_tg = df_tg.dropna(subset=['salary_from']).reset_index(drop=True)
df_tg
```

```python
def grade(text):
  text = text.lower()
  if any(w in text for w in ['senior', 'сеньор', 'сеньора', 'старший', 'старшего']):
    return 'senior'
  elif any(w in text for w in ['middle', 'мидл', 'миддл', 'мидла', 'миддла']):
    return 'middle'
  elif any(w in text for w in ['junior', 'джун', 'джуна', 'младший', 'младшего']):
    return 'junior'
  elif any(w in text for w in ['lead', 'team lead', 'teamlead', 'тимлид', 'тимлида']):
    return 'lead'
  else:
    return None
```

```python
def technologies(text):
  found = []
  text_lower = text.lower()
  for tech in tech_keywords:
    tech_lower = tech.lower()
    if tech_lower in text_lower:
      found.append(tech)
  return ', '.join(found) if found else None
```
