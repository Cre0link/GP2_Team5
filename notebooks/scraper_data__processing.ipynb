```python
def split_salary(s):
  s = str(s).replace('\xa0', ' ')
  nums = re.findall(r'\d[\d ]*', s)
  numbers = []
  for part in nums:
    part = part.replace(' ', '')
    numbers.append(float(part))
  if len(numbers) == 0:
    return np.nan, np.nan, np.nan
  if len(numbers) == 1:
    salary_from = numbers[0]
    salary_to = np.nan
    salary_mid = numbers[0]
    return salary_from, salary_to, salary_mid
  salary_from = min(numbers[0], numbers[1])
  salary_to = max(numbers[0], numbers[1])
  salary_mid = (salary_from + salary_to)/2

  return float(salary_from), float(salary_to), float(salary_mid)
```

```python
sal_from = []
sal_to = []
sal_mid = []

for s in df_hh['salary']:
  salary_from, salary_to, salary_mid = split_salary(s)
  sal_from.append(salary_from)
  sal_to.append(salary_to)
  sal_mid.append(salary_mid)

df_hh['salary_from'] = sal_from
df_hh['salary_to'] = sal_to
df_hh['salary_mid'] = sal_mid
df_hh
```

```python
def grade(text):
  text = text.lower()
  if any(w in text for w in ['senior', 'сеньор', 'сеньора', 'старший', 'старшего']):
    return 'senior'
  elif any(w in text for w in ['middle', 'мидл', 'миддл', 'мидла', 'миддла']):
    return 'middle'
  elif any(w in text for w in ['junior', 'джун', 'джуна', 'младший', 'младшего']):
    return 'junior'
  elif any(w in text for w in ['lead', 'team lead', 'teamlead', 'тимлид', 'тимлида']):
    return 'lead'
  else:
    return None
```

```python
df_hh['grade'] = df_hh['text'].apply(grade)
df_hh['grade'].value_counts(dropna=False)
```

```python
def work_format(text):
  text = text.lower()
  if any(w in text for w in ['удалён', 'удален', 'удалёнка', 'удаленка', 'remote']):
    return 'remote'
  elif any(w in text for w in ['гибрид', 'hybrid']):
    return 'hybrid'
  elif any(w in text for w in ['офис', 'очно', 'office']):
    return 'office'
  else:
    return None
```

```python
def technologies(text):
  found = []
  text_lower = text.lower()
  for tech in tech_keywords:
    tech_lower = tech.lower()
    if tech_lower in text_lower:
      found.append(tech)
  return ', '.join(found) if found else None
```
